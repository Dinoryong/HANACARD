```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import re
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('C:\data\CardData_new.csv', encoding='cp949')
df.head()
```

```python
def month_to_number(month):
    month_dict = {
        '21.Active_연속': 1, '34.휴면_9M+': 0, '31.휴면_3M': 0, '33.휴면_9M-': 0,
        '22.Active_간헐': 1, '32.휴면_6M-': 0
    }
    return month_dict.get(month, month)  # 딕셔너리에 없는 값이면 그대로 반환

df['고객구분'] = df['고객구분'].apply(month_to_number)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder

# 데이터 준비
X = df[['탈회횟수_발급1년이내','_1순위카드이용금액','_1순위카드이용건수','카드이용한도금액','이용후경과월_신용']]
y = df['고객구분']



# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

# RandomForestClassifier 모델 생성 및 학습
rf_model = RandomForestClassifier(n_estimators=100, random_state=13)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# 변수 중요도 출력
feature_importances = pd.DataFrame(rf_model.feature_importances_,
                                   index=X_train.columns,
                                   columns=['importance']).sort_values('importance', ascending=False)

print(feature_importances)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# 데이터 불러오기
df = pd.read_csv('C:\data\CardData_new.csv', encoding='cp949')

def month_to_number(month):
    month_dict = {
        '21.Active_연속': 1, '34.휴면_9M+': 0, '31.휴면_3M': 0, '33.휴면_9M-': 0,
        '22.Active_간헐': 1, '32.휴면_6M-': 0
    }
    return month_dict.get(month, month)

# 훈련 데이터와 테스트 데이터로 분할
train_data = df[df['고객구분'] == 0]
test_data = df[df['고객구분'] == 1]

# 훈련 데이터에서 필요한 컬럼 선택
selected_columns = ['탈회여부', '남녀구분코드', '연령', '소지여부_신용', '입회일자_신용', '입회경과개월수_신용', '회원여부_연체']
X_train = train_data[selected_columns]
y_train = train_data['고객구분']

# 랜덤 포레스트 모델 생성 및 학습
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# 테스트 데이터에서 필요한 컬럼 선택
X_test = test_data[selected_columns]
y_test = test_data['고객구분']

# 예측
y_pred = rf_model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print(f'모델의 정확도: {accuracy:.2f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# 데이터 불러오기
df = pd.read_csv('C:\data\CardData_new.csv', encoding='cp949')

# 고객구분을 기준으로 데이터 분할
train_data = df[df['고객구분'] == 0]
test_data = df[df['고객구분'] == 1]

# 훈련 데이터와 테스트 데이터에서 '고객구분' 칼럼 제거
X_train = train_data.drop('고객구분', axis=1)
X_test = test_data.drop('고객구분', axis=1)

# 훈련 및 테스트 데이터의 목표 변수 설정
y_train = train_data['고객구분']
y_test = test_data['고객구분']

# RandomForestClassifier 모델 생성 및 학습
rf_model = RandomForestClassifier(n_estimators=100, random_state=13)
rf_model.fit(X_train, y_train)

# 모델 평가
accuracy = rf_model.score(X_test, y_test)
print(f'모델의 정확도: {accuracy:.2f}')
```
