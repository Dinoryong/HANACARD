```python
df3 = df3[df3['서비스코드명'] != '_']
```

```python
df3['서비스코드명'].unique()
```

```python
benefit_mapping = {
    '온라인쇼핑할인': 'a',
    '주유 할인': 'b',
    '놀이공원할인': 'c',
    '아파트할인': 'd',
    '유통/마트 할인': 'e',
    '교육관련할인': 'f',
    '통신할인': 'g',
    '문화생활할인': 'h',
    '외식할인': 'i',
    '기타 원금할인': 'j',
    '커피할인': 'k',
    '교통할인': 'l',
    '전가맹점할인': 'm',
    '가전할인': 'n',
    '기타': 'o',
    '기타 포인트': 'p',
    '플래티늄서비스': 'q',
    '자동차할인': 'r',
    '세이브포인트': 's',
    '캐쉬백서비스': 't',
    '대한항공 항공사 마일리지': 'u',
    '아시아나 항공사 마일리지': 'v',
}
```

```python
# '서비스코드명' 열의 값을 변환하여 새로운 열 추가
df3['알파벳코드'] = df3['서비스코드명'].map(benefit_mapping)
```

```python
# 중복된 혜택을 가진 카드들을 묶어서 출력하는 함수
def print_duplicate_benefits(df):
    grouped = df.groupby('알파벳코드')['_1순위카드ID'].apply(list)
    
    for benefit_code, card_ids in grouped.items():
        if len(card_ids) > 1:
            print("=" * 50)
            print("Card ID:", ', '.join(map(str, card_ids)))
            for code, benefit in benefit_mapping.items():
                if benefit == benefit_code:
                    print(f"{benefit} ({code})")
                    break
            print("=" * 50)

# 중복된 혜택을 가진 카드들을 출력
print_duplicate_benefits(df3)
```

```python
# 각 카드번호에 대한 혜택을 리스트로 묶어주는 딕셔너리 생성
card_benefits_dict = {}

for index, row in df3.iterrows():
    card_id = row['_1순위카드ID']
    benefit_code = row['알파벳코드']
    
    if card_id not in card_benefits_dict:
        card_benefits_dict[card_id] = [benefit_code]
    else:
        card_benefits_dict[card_id].append(benefit_code)

# 결과 출력
for card_id, benefits in card_benefits_dict.items():
    print(f"카드 ID: {card_id} = {benefits}")
```

```python
from collections import defaultdict

# 각 혜택에 대한 카드 ID를 defaultdict로 묶어주기
benefit_to_card_dict = defaultdict(list)

for card_id, benefits in card_benefits_dict.items():
    benefit_key = tuple(sorted(benefits))  # 리스트를 정렬하여 튜플로 변환하여 키로 사용
    benefit_to_card_dict[benefit_key].append(card_id)

# 같은 혜택을 가진 카드 ID 묶음 출력
cnt=0
for benefits, card_ids in benefit_to_card_dict.items():
    card_id_strings = [str(card_id) for card_id in card_ids]  # 카드 ID를 문자열로 변환
    cnt+=1
    print(f"혜택 {benefits}: 카드ID {', '.join(card_id_strings)}")
print(cnt)
```

```python
# 같은 혜택을 가진 카드 ID 묶음 출력 및 갯수 세기
for benefits, card_ids in benefit_to_card_dict.items():
    card_id_strings = [str(card_id) for card_id in card_ids]  # 카드 ID를 문자열로 변환
    print(f"혜택 {benefits}: 카드ID {', '.join(card_id_strings)} (갯수: {len(card_ids)})")
```
