```python
for column in dfnew:
    if np.issubdtype(df[column].dtype, np.number):  # 숫자 타입인 컬럼만 선택
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        print(f"Column: {column}, Number of outliers: {len(outliers)}")
```

```python
for column in dfnew:
    if np.issubdtype(df[column].dtype, np.number):  # 숫자 타입인 컬럼만 선택
        valid_data = df[df[column] > 0]  # 0과 음수를 제외한 데이터 선택
        Q1 = valid_data[column].quantile(0.25)
        Q3 = valid_data[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers0 = valid_data[(valid_data[column] < lower_bound) | (valid_data[column] > upper_bound)]

        print(f"Column: {column}, Number of outliers: {len(outliers0)}")
```

```python
for column in dfnew:
    if np.issubdtype(df[column].dtype, np.number):  # 숫자 타입인 컬럼만 선택
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        valid_data = df[df[column] > 0]  # 0과 음수를 제외한 데이터 선택
        Q1_valid = valid_data[column].quantile(0.25)
        Q3_valid = valid_data[column].quantile(0.75)
        IQR_valid = Q3_valid - Q1_valid

        lower_bound_valid = Q1_valid - 1.5 * IQR_valid
        upper_bound_valid = Q3_valid + 1.5 * IQR_valid

        outliers0 = valid_data[(valid_data[column] < lower_bound_valid) | (valid_data[column] > upper_bound_valid)]

        outliers_diff = len(outliers0) - len(outliers)
        print(f"Column: {column}, Difference in Number of Outliers: {outliers_diff}, {len(outliers0)}")
        
```
