```python
#칼럼별 boxplot그리는 함수
def boxplot(dfnew):
    num_columns = len(dfnew.columns)
    num_rows = (num_columns + 1) // 2  

    plt.figure(figsize=(12, 6 * num_rows))

    for i, column in enumerate(dfnew.columns):
        plt.subplot(num_rows, 5, i + 1)
        dfnew[column].plot(kind='box')

    plt.tight_layout()  
    plt.show()
```

```python
boxplot(dfnew)
```

```python
#칼럼별 displot그리는 함수
def displot(dfnew):
    num_columns = len(dfnew.columns)
    num_rows = (num_columns + 1) // 2  

    plt.figure(figsize=(12, 6 * num_rows))

    for i, column in enumerate(dfnew.columns):
        plt.subplot(num_rows, 5, i + 1)
        dfnew[column].plot(kind='kde')

    plt.tight_layout()  
    plt.show()
```

```python
displot(dfnew)
```

```python
#이상치 확인 코드
def outlier(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower= q1 - 1.5 * iqr
    upper= q3 + 1.5 * iqr
    outliers = column[(column < lower) | (column > upper)]
    return outliers
```

```python
#칼럼별 이상치 갯수 반환
for column in dfnew.columns:
    outliers = outlier(dfnew[column])
    if not outliers.empty:
        print(column)
        print(len(outliers))
```
