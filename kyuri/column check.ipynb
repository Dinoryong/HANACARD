# 칼럼별 boxplot그리는 함수

```python
#칼럼별 boxplot그리는 함수
def boxplot(dfnew):
    num_columns = len(dfnew.columns)
    num_rows = (num_columns + 1) // 2  

    plt.figure(figsize=(12, 6 * num_rows))

    for i, column in enumerate(dfnew.columns):
        plt.subplot(num_rows, 5, i + 1)
        dfnew[column].plot(kind='box')

    plt.tight_layout()  
    plt.show()
```

```python
boxplot(dfnew)
```

# 이상치 확인 함수

```python
#이상치 확인 코드
def outlier(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower= q1 - 1.5 * iqr
    upper= q3 + 1.5 * iqr
    outliers = column[(column < lower) | (column > upper)]
    return outliers
```

```python
#칼럼별 이상치 갯수 반환
for column in dfnew.columns:
    outliers = outlier(dfnew[column])
    if not outliers.empty:
        print(column)
        print(len(outliers))
```

# 결측치 확인 함수

```python
def null(df):
    missing_values = df.isnull().sum()
    total_values = df.shape[0]
    missing_percentage = (missing_values / total_values) * 100
    result_df = pd.DataFrame({
        'Missing Values': missing_values,
        'Missing Percentage': missing_percentage
    })
    return result_df
```

```python
result = null(dfnew)
print(result)
```

```python
for column in dfnew:
    Q1 = dfnew[column].quantile(0.25)
    Q3 = dfnew[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = dfnew[(dfnew[column] < lower_bound) | (dfnew[column] > upper_bound)]

    print(f"Column: {column}, Number of outliers: {len(outliers)}")
```

```python
#0제외하고 이상치 확인
for column in dfnew:
    nonzero_values = df[column][df[column] > 0]  # 0을 제외한 값 선택

    q3 = nonzero_values.quantile(0.75)
    q1 = nonzero_values.quantile(0.25)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = nonzero_values[(nonzero_values < lower_bound) | (nonzero_values > upper_bound)]

    print(f"Column: {column}, Number of outliers: {len(outliers)}")
```
